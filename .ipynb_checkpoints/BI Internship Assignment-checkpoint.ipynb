```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the data file 

df1 = pd.read_csv('BI-Intern first_purchases.csv')
df1.head()
```

```python
df2 = pd.read_csv('BI-Intern purchases.csv')
df2.head()
```

```python
# Divide the two data files by product line

df3 = df1[df1['First Purchase Product Line']=="Restaurant"]
df4 = df1[df1['First Purchase Product Line']=='Retail store']
df5 = df2[df2['Product line']=='Restaurant']
df6 = df2[df2['Product line']=='Retail store']
```

```python
# Exclude the users whose first purchase were before May 2020

df7 = df5[df5['User ID'].isin(df3['User ID'])]
df8 = df6[df6['User ID'].isin(df4['User ID'])]
```

```python
# Save the files as csv

df3.to_csv('Restaurant_New user.csv',index=False)
df4.to_csv('Retail_New user.csv',index=False)
df7.to_csv('Restaurant_New user purchase.csv',index=False)
df8.to_csv('Retail_New user purchase.csv',index=False)
```
